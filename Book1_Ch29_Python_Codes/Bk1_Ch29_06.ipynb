```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

image.png

```python
# 导入鸢尾花数据
X,y = load_iris(return_X_y=True)
# 拆分鸢尾花数据集为训练集和测试集
```

##划分数据集

参数 test_size 为 0.2，表示将数据的 20% 作为测试集，剩余 80% 作为训练集。

这个参数决
定了训练集和测试集的划分比例。


X_train 为训练集的特征矩阵，包含用于训练机器学习模型的特征数据。


X_test 为测试集的特征矩阵，包含用于评估模型性能的特征数据。


y_train 为训练集的标签向量，包含训练集样本对应的目标值。


y_test 为测试集的标签向量，包含测试集样本对应的目标值。


```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2)
```

#自定义可视化函数

创建一个包含 1 行、2 列的子图布局。gridspec_kw={'width_ratios': [4, 1]} 参数
用于控制每个子图的宽度比例，这里设置了第一个子图的宽度为第二个子图的 4 倍。


```python
# 自定义可视化函数
def visualize(df):
    fig, axs = plt.subplots(1, 2,
        gridspec_kw={'width_ratios': [4, 1]})

    # 绘制热图（取所有数据但不包括最后一行数据）
    # ax = axs[0] 作用第一个子图
    sns.heatmap(df.iloc[:,0:-1],
                cmap='RdYlBu_r', yticklabels = False,
                cbar=False, ax = axs[0])
    # 绘制热图（取最后一行数据）
    # ax = axs[0] 作用第二个子图
    sns.heatmap(df.iloc[:,[-1]],
                cmap='Set3', yticklabels = False,
                cbar=False, ax = axs[1])
```

```python
# 转化为Pandas DataFrame
columns = ['Sepal length, X1', 'Sepal width, X2',
           'Petal length, X3', 'Petal width, X4',
           'Species']

#np.c_[X,y]将特征数据与标签数据按照列拼接
# 重命名列标签
df_full = pd.DataFrame(np.c_[X,y],
                       columns = columns)
```

```python
df_full
```

```python
visualize(df_full)
```

# 训练集

```python
# 训练集
# np.c_[X_train, y_train]部分数据（训练数据）
df_train = pd.DataFrame(np.c_[X_train, y_train],
                        columns = columns)
```

```python
df_train
```

```python
visualize(df_train)
```

# 测试集

```python
# 测试集
df_test = pd.DataFrame(np.c_[X_test, y_test],
                       columns = columns)
```

```python
visualize(df_test)
```
