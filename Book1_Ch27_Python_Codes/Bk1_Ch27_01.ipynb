```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from statsmodels.graphics.plot_grids import scatter_ellipse # 散点椭圆
```

```python
# 导入鸢尾花数据
data_raw = sns.load_dataset('iris')
labels = ['Sepal length','Sepal width',
          'Petal length','Petal width']
```

```python
data_raw
```

可能的输出示例
假设有 3 个变量（A、B、C），输出可能是 3×3 的子图网格：

对角线：变量名或密度图

非对角线：散点图 + 椭圆

椭圆越扁 → 两变量相关性越强（接近线性关系）

椭圆接近圆 → 两变量相关性弱

椭圆倾斜方向 → 正相关/负相关方向

scatter_ellipse 函数是 statsmodels.graphics.plot_grids 模块的一部分，用于创
建带有椭圆表示置信区间的散点图。简单来说，scatter_ellipse 函数在基本散点图的基础上添加
了椭圆，用于展示样本数据的置信区间。


绘制平面散点图 + 椭圆。scatter_ellipse 函数中，level (默认 0.9) 是一个可选参
数，用于控制绘制椭圆时表示置信区间 (confidence interval) 的置信水平 (confidence
level)。


```python
fig = plt.figure(figsize=(8,8))
scatter_ellipse(data_raw.iloc[:,:-1], #选取数据；但不包括最后一列
                varnames=labels, fig=fig)
# fig.savefig('散点 + 椭圆.svg', format='svg')
```

绘制平面散点图 + 椭圆。scatter_ellipse 函数中，level (默认 0.9) 是一个可选参
数，用于控制绘制椭圆时表示置信区间 (confidence interval) 的置信水平 (confidence
level)。


```python
for s_idx in data_raw.species.unique():
    data= data_raw.loc[data_raw.species == s_idx].iloc[:,:-1]
    fig = plt.figure(figsize=(8,8))
    scatter_ellipse(data, varnames=labels, fig=fig)
    # fig.savefig('散点 + 椭圆 ' + s_idx + '.svg', format='svg')
  # 包括每个种类的 散点椭圆图
```
