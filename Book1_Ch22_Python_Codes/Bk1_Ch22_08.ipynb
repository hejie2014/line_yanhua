# 利用stack() 将宽格式转换为长格式
Book_1《编程不难》 | 鸢尾花书：从加减乘除到机器学习  

```python
import pandas as pd
import numpy as np
```

```python
student_ids = [1, 2, 3, 4]
subjects = ['Art', 'Math', 'Science']
np.random.seed(0)
# 使用随机数生成成绩数据
scores = np.random.randint(3, 6,
            size=(len(student_ids),len(subjects)))
```

```python
# 创建数据帧
df = pd.DataFrame(scores, index=student_ids,
                  columns=subjects)
df
```

```python
df.columns.names
```

```python
df.index.names
```

```python
df.index
```

```python
df.columns
```

```python
# 修改行列名称
df.columns.names = ['Subject']
df.index.names = ['Student ID']
df
```

```python
df.stack()  # 将Subject转为行
```

```python
df.stack().reset_index()  # 将行索引转为列索引
```

```python
# 将长格式转化为宽格式
df.stack().reset_index().rename(columns={0: 'Final'}) # 将0重命名为Final
```
