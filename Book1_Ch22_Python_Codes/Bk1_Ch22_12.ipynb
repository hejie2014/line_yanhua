# 鸢尾花数据帧groupby(['species']) 计算统计量
Book_1《编程不难》 | 鸢尾花书：从加减乘除到机器学习  

```python
import seaborn as sns
import pandas as pd
```

```python
iris_df = sns.load_dataset("iris")
# 从Seaborn中导入鸢尾花数据帧
```

```python
# 分组计算统计量
groupby_mean = iris_df.groupby(['species']).mean()
# 按照种类进行分类；并计算每个特性的均值
groupby_mean
```

```python
groupby_std  = iris_df.groupby(['species']).std() #标准差
groupby_std
```

```python
groupby_var  = iris_df.groupby(['species']).var() #方差
groupby_var
```

核心区别一句话

函数	含义	数值范围	是否受量纲影响

cov()	协方差：衡量两个变量同向变化的绝对程度	(-∞, +∞)	✅ 受影响

corr()	相关系数：衡量两个变量线性关系的强度和方向	[-1, 1]	❌ 不受影响

```python
groupby_cov  = iris_df.groupby(['species']).cov() #协方差
groupby_cov
```

```python
groupby_corr = iris_df.groupby(['species']).corr()#相关系数
groupby_corr
```
