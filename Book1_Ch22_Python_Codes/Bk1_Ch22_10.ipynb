```python
import pandas as pd
```

```python
data = {
    ('A', 'Art'):     [4, 3, 5, 4],
    ('A', 'Math'):    [3, 4, 5, 3],
    ('A', 'Science'): [5, 4, 3, 4],
    ('B', 'Art'):     [3, 4, 5, 4],
    ('B', 'Math'):    [4, 5, 3, 3],
    ('B', 'Science'): [5, 3, 4, 5]}
    # ('A', 'Art')二级列索引
```

```python
# 创建多层行标签数据帧
df = pd.DataFrame(data, index=[1, 2, 3, 4])
df
```

```python
# 添加行标签名称
df.columns.names = ['Class', 'Subject'] # 使用list多级索引
df.index.names = ['Student ID']
df
```

```python
# 选择 'Class' 进行 stack() 操作
stacked_df = df.stack(level='Class') # 指定需要转换的列标签
stacked_df
# stacked_df = df.stack(level=0)
```

```python
stacked_df = df.stack(level=[0,1])
```
