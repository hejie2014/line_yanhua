# 利用merge() 完成左侧独有、右侧独有、全外独有
Book_1《编程不难》 | 鸢尾花书：从加减乘除到机器学习  

```python
import pandas as pd
```

```python
# 创建两个数据帧
left_data = {
    'M': [ 1,   2,   3],
    'X': ['a', 'b', 'c']}
left_df = pd.DataFrame(left_data)
left_df
```

```python
right_data = {
    'X': ['b', 'c', 'd'],
    'N': [ 22,  33,  44]}
right_df = pd.DataFrame(right_data)
right_df
```


首先利用 merge() 完成左连接合并。在 pandas 的 merge() 方法中，
indicator 参数用于指定是否添加一个特殊的列，该列记录了每行的合并方式。
这个特殊的列名可以通过 indicator 参数进行自定义，默认为 "_merge"。"_merge" 列可
以取三个值：
"left_only": 表示该行只在左边的 DataFrame 中存在，即左连接中独有的行。
"right_only": 表示该行只在右边的 DataFrame 中存在，即右连接中独有的行。
"both": 表示该行在两个 DataFrame 中都存在，即连接方式中共有的行。
在
b
中，通过设定筛选条件，left_exl['_merge'] == 'left_only'，我们可以保留合并后
的“左侧独有”行。

```python
# LEFT EXCLUSIVE
left_exl = left_df.merge(right_df,
                         on='X',
                         how='left',
                         indicator=True)
```

```python
left_exl  #以left数据帧为基准； a为left独有；b c：共有
```

```python
left_exl = left_exl[
    left_exl['_merge'] == 'left_only'].drop(
    columns=['_merge'])
left_exl
# 选取 ——merge为left_only的行数据； 删除_merge列
```

```python
# RIGHT EXCLUSIVE
right_exl = left_df.merge(right_df,
                          on='X',
                          how='right',
                          indicator=True)
```

```python
right_exl
```

```python
right_exl = right_exl[
    right_exl['_merge'] == 'right_only'].drop(
    columns=['_merge'])
right_exl
```

```python
# FULL OUTER EXCLUSIVE
outer_exl = left_df.merge(right_df,
                          on='X',
                          how='outer',
                          indicator=True)
```

```python
outer_exl
```

```python
outer_exl = outer_exl[
    outer_exl['_merge'] != 'both'].drop(
    columns=['_merge'])
outer_exl
```
