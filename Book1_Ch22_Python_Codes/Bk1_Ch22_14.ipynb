# 鸢尾花数据帧使用apply() 自定义函数，对于特定一列
Book_1《编程不难》 | 鸢尾花书：从加减乘除到机器学习  

```python
import seaborn as sns
import pandas as pd
```

```python
iris_df = sns.load_dataset("iris")
# 从Seaborn中导入鸢尾花数据帧
iris_df
```

```python
# 定义函数将花萼长度映射为等级
def map_fnc(sepal_length):
    if sepal_length < 5:
        return 'D'
    elif 5 <= sepal_length < 6:
        return 'C'
    elif 6 <= sepal_length < 7:
        return 'B'
    else:
        return 'A'
```

利用 apply() 将自定义函数用在数据帧 iris_df['sepal_length'] 上。


apply() 方法可以接受一个函数作为参数，这个函数将会被应用到 DataFrame 的每一行或每
一列上。这个函数可以是 Pandas 中已经定义好的函数，可以是自定义函数，也可以是匿名 lambda
函数。


```python
# 使用 apply 函数将 sepal_length 映射为等级并添加新列
iris_df['ctg'] = iris_df['sepal_length'].apply(map_fnc)
iris_df
```
