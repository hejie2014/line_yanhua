# 鸢尾花数据帧使用apply() 匿名lambda函数，对于特定一列
Book_1《编程不难》 | 鸢尾花书：从加减乘除到机器学习  

```python
import seaborn as sns
import pandas as pd
```

```python
iris_df = sns.load_dataset("iris")
# 从Seaborn中导入鸢尾花数据帧
iris_df
```

```python
# 使用apply() 和lambda函数计算每个类别中最小的花瓣宽度
iris_df.groupby('species')['sepal_length'].apply(
    lambda x: x.min())
# iris_df.groupby('species')['sepal_length'].min()
```

```python
iris_df.groupby('species')
```

```python
iris_df.groupby('species')['sepal_length']
```

```python
iris_df.groupby('species')['sepal_length'].min()
```

```python
iris_df.groupby('species')['sepal_length'].count()
```

```python
# 示例1：计算每个品种的花萼长度范围（最大值-最小值）
iris_df.groupby('species')['sepal_length'].apply(
    lambda x: x.max() - x.min()
)
```

```python
# 示例2：计算每个品种中，有多少样本的花萼长度大于该品种平均值
import numpy as np
iris_df.groupby('species')['sepal_length'].apply(
    lambda x: (x > x.mean()).sum()
)
```

```python
# 示例3：对每个品种的数据进行自定义复杂处理
iris_df.groupby('species')['sepal_length'].apply(
    lambda x: np.percentile(x, [25, 50, 75])  # 计算四分位数
)
```

```python
import time

# 用apply + lambda
start = time.time()
result1 = iris_df.groupby('species')['sepal_length'].apply(lambda x: x.min())
print("apply耗时:", time.time() - start)

# 直接用min
start = time.time()
result2 = iris_df.groupby('species')['sepal_length'].min()
print("min耗时:", time.time() - start)
```
