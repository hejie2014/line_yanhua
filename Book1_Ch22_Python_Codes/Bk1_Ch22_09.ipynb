# 利用melt() 将宽格式转换为长格式
Book_1《编程不难》 | 鸢尾花书：从加减乘除到机器学习  

```python
import pandas as pd
```

```python
data = {
    'Student ID': ['1', '2', '3', '4'],
    'Art':        [4, 3, 5, 4],
    'Math':       [3, 4, 5, 3],
    'Science':    [5, 4, 3, 4]}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.columns
```

```python
df.columns.names = ['Subject'] #
df
```

```python
df.columns
```

melt() 需要指定 id_vars 参数，表示保持不变的列，同时还可以选择 value_vars 参数来指
定哪些列需要被转换。

```python
melted_df = df.melt(id_vars='Student ID',
                    var_name='Subject', #将需要转化的列的标签作为subject
                    value_vars=['Art','Math','Science'],
                    value_name='Score')
melted_df
```
