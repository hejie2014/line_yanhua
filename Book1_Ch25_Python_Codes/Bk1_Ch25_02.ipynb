## 正定性

正定性 (positive definiteness) 是线性代数、优化方法、机器学习重要的数学概念。表 4
用一组 2 × 2 矩阵 A2×2 介绍正定性，下面具体来看。

矩阵 A2×2 是正定 (positive definite)，意味着 f(x) = x
T @ A2×2 @ x 是个开口朝上的抛物
面，形状像是碗。除了 (0, 0)，f(x) = x
T @ A2×2 @ x 均大于 0。(0, 0) 为最小值。

矩阵 A2×2 是半正定 (positive semi-definite)，意味着 f(x) = x
T @ A2×2 @ x 是个开口朝
上的山谷面。除了 (0, 0)，f(x) = x
T @ A2×2 @ x 均大于等于 0。山谷的谷底都是极小值。

矩阵 A2×2 是负定 (negative definite)，意味着 f(x) = x
T @ A2×2 @ x 是个开口朝下的抛物
面。除了 (0, 0)，f(x) = x
T @ A2×2 @ x 均小于 0。(0, 0) 为最大值。

矩阵 A2×2 是半负定 (negative semi-definite)，意味着 f(x) = x
T @ A2×2 @ x 是个开口朝
下的山脊面。除了 (0, 0)，f(x) = x
T @ A2×2 @ x 均小于等于 0。山脊的顶端都是极大值。

矩阵 A2×2 不定 (indefinite)，意味着 f(x) = x
T @ A2×2 @ x 是个马鞍面，(0, 0) 为鞍点。
f(x) = x
T @ A2×2 @ x 符号不定。

```python
from sympy import Matrix, symbols
from sympy.plotting import plot3d
x1, x2 = symbols('x1 x2')
x = Matrix([x1, x2])
```

### 正定

$${\displaystyle M{\text{ positive-definite}}\quad \iff \quad \mathbf {x} ^{\textsf {T}}M\mathbf {x} >0{\text{ for all }}\mathbf {x} \in \mathbb {R} ^{n}\setminus \{\mathbf {0} \}}$$

```python
A = Matrix([[1, 0], [0, 1]])
A.is_positive_definite # 是否是正定矩阵
```

```python
x.T*A*x
```

```python
f_x = x.T*A*x
f_x
```

```python
print(type(f_x))
```

```python
f_x = x.T*A*x
f_x = f_x[0]
f_x
```

x.T * A * x 计算出来是一个 1×1 的矩阵（标量），而不是纯数字。

[0, 0] 就是取出这个矩阵中的唯一元素（第 0 行第 0 列），得到一个普通的标量表达式，这样才能被 plot3d 正确解析。

```python
p = plot3d((x.T*A*x)[0, 0], (x1, -1, 1), (x2, -1, 1))
```

### 半正定

$${\displaystyle M{\text{ positive semi-definite}}\quad \iff \quad \mathbf {x} ^{\textsf {T}}M\mathbf {x} \geq 0{\text{ for all }}\mathbf {x} \in \mathbb {R} ^{n}}$$

矩阵 A2×2 是半正定 (positive semi-definite)，意味着 f(x) = x
T @ A2×2 @ x 是个开口朝
上的山谷面。除了 (0, 0)，f(x) = x
T @ A2×2 @ x 均大于等于 0。山谷的谷底都是极小值。


```python
A = Matrix([[1, -1], [-1, 1]])
A.is_positive_definite #是否为正定 (positive definite)，
```

```python
A.is_positive_semidefinite #是否为半正定 (positive semi-definite)，
```

### 负定

$$ {\displaystyle M{\text{ negative-definite}}\quad \iff \quad \mathbf {x} ^{\textsf {T}}M\mathbf {x} <0{\text{ for all }}\mathbf {x} \in \mathbb {R} ^{n}\setminus \{\mathbf {0} \}} $$

矩阵 A2×2 是负定 (negative definite)，意味着 f(x) = x
T @ A2×2 @ x 是个开口朝下的抛物
面。除了 (0, 0)，f(x) = x
T @ A2×2 @ x 均小于 0。(0, 0) 为最大值。


```python
A = Matrix([[-1, 0], [0, -1]])
A.is_negative_definite
```

```python
A.is_negative_semidefinite
```

```python
A.is_indefinite
```

### 半负定

$$ {\displaystyle M{\text{ negative semi-definite}}\quad \iff \quad \mathbf {x} ^{\textsf {T}}M\mathbf {x} \leq 0{\text{ for all }}\mathbf {x} \in \mathbb {R} ^{n}} $$

矩阵 A2×2 是半负定 (negative semi-definite)，意味着 f(x) = x
T @ A2×2 @ x 是个开口朝
下的山脊面。除了 (0, 0)，f(x) = x
T @ A2×2 @ x 均小于等于 0。山脊的顶端都是极大值。
